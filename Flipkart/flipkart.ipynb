````python
# Importing all the Important Libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
from prettytable import PrettyTable
````

<h2>Fetch HTML Data From the URL Using BeautifulSoup</h2>

````python
# Store URL into Website Variable, For e.g.:- I have Taken Flipkart Shopping for Data Scrapping !!!
website_url = "https://www.flipkart.com/search?q=iphone%2012&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
````

````python
# Now Request Web Page and fetch HTML Data inside it for Scrapping Procedure
web_request = requests.get(website_url)
fetch_url = urlopen(website_url)
page_html = fetch_url.read()
````

````python
# Parse HTML ans CSS Content to Soup Variable
soup = BeautifulSoup(page_html, "html.parser")
````

<h3>Print HTML Content in Pretiffy Format</h3>

````python
# Print Code Using Prettify Format
print(soup.prettify())
````

<h2>Find Product Name and Product Price using findAll() function and Strip Down all the Data Using strip() Function !</h2>

````python
# Fetch Code for Product Name and other information
product_name = soup.findAll('div', attrs = {'class':'_4rR01T'})
print("Length of the Poduct Data:- ", len(product_name))
print("")
print(product_name)
````

````python
# Print Product Name in the Structured Format Using strip() Function
for i in range(0, 24):
    print(product_name[i].text.strip())
````

````python
# Fetch Code for Product Price and other information
product_price = soup.findAll('div', attrs = {'class':'_30jeq3 _1_WHN1'})
print("Length of the Poduct Data:- ", len(product_price))
print("")
print(product_price)
````

````python
# Print Product Price in the Structured Format Using strip() Function
for i in range(0, 24):
    print(product_price[i].text.strip())
````

````python
# Print Product Nme and Price in the Structured Format Using strip() Function
print("Flipkart iPhone Price List Data:- \n")
for i in range(0, 24):
    print("Product Name:- ", product_name[i].text.strip(), "\nPrice:- ", product_price[i].text.strip(), "\n")
````

````python
# Fetch Code for Product Specification and other information
product_specs = soup.findAll('li', attrs = {'class':'rgWa7D'})
print("Lenth of Product Information:- ", len(product_specs))
print("")
print(product_specs)
````

````python
# Print Product Specification in the Structured Format Using strip() Function
for i in range(0, 189):
    if i%8 == 7:
        print("")
    else:
        pass
    print(product_specs[i].text.strip())
    
````
